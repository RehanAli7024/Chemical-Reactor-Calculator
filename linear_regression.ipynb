```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pylab as pl
```

```python
df = pd.read_csv("/content/FuelConsumptionCo2.csv")

# take a look at the dataset
df.head()
```

```python
df.describe()
```

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9)
```

```python
viz = cdf[['CYLINDERS','ENGINESIZE','CO2EMISSIONS','FUELCONSUMPTION_COMB']]
viz.hist()
plt.show()
```

```python
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("FUELCONSUMPTION_COMB")
plt.ylabel("Emission")
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color = 'red')
plt.xlabel("Engine Size")
plt.ylabel("Emission")
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (train_x, train_y)
print("slope", regr.coef_)
print("intercept", regr.intercept_)
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Engine size")
plt.ylabel("Emission")
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y , test_y_) )
```

```python
rnd = np.random.rand(len(cdf)) < 0.8
trainset = cdf[rnd]
testset = cdf[~rnd]
from sklearn import linear_model
lg = linear_model.LinearRegression()
trainset_x = np.asanyarray(trainset[['FUELCONSUMPTION_COMB']])
trainset_y = np.asanyarray(trainset[['CO2EMISSIONS']])
lg.fit(trainset_x, trainset_y)
print("slope", lg.coef_)
print("intercept", lg.intercept_)
```

```python
from sklearn.metrics import r2_score
testsetx = np.asanyarray(testset[['FUELCONSUMPTION_COMB']])
testsety = np.asanyarray(testset[['CO2EMISSIONS']])
testsety_ = lg.predict(testsetx)
print("Mean absolute error: %.2f" % np.mean(np.absolute(testsety_ - testsety)))
print("Residual sum of squares (MSE): %.2f" % np.mean((testsety_ - testsety) ** 2))
print("R2-score: %.2f" % r2_score(testsety , testsety_) )
```
